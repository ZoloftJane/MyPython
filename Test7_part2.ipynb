```python
# Выведите на экран таблицу умножения чисел от 1 до 9.
a=1
while a<10:
    b=1
    while b<10:
        c=a*b
        print(c, end=" ")
        b+=1
    print(" ")
    a+=1
```

```python
# Выведите на экран таблицу сложения чисел от 1 до 9, т.е. каждый элемент таблицы это значение суммы индекса строки и столбца.
a=1
while a<10:
    b=1
    while b<10:
        c=a+b
        print(c, end=" ")
        b+=1
    print(" ")
    a+=1
```

```python
# Выведите на экран таблицу степеней числа a в степени b, где a и b - числа от 1 до 5, 
# т.е. каждый элемент таблицы это значение индекса строки, возведенное в степень равное индексу столбца
a=1
while a<=5:
    b=1
    while b<=5:
        c=a**b
        print(c, end=" ")
        b+=1
    print(" ")
    a+=1
```

```python
# Выведите на экран таблицу деления числа a на число b, где a и b - числа от 1 до 5, 
# т.е. каждый элемент таблицы это значение частного от деления индекса строки на индекс столбца 
# с точностью 2 знака после запятой. Для форматирования вывода дробного числа используйте функцию round(x, 2) 
# Эта функция принимает на вход число x и количество знаков после запятой, которые нужно оставить при выводе. 
a=1
while a<=5:
    b=1
    while b<=5:
        c=a/b
        print(round(c, 2), end=" ")
        b+=1
    print(" ")
    a+=1
```

```python
# Программа получает на вход натуральное число N (1 ≤ N ≤100). 
# Выведите на экран таблицу размером NxN, элементы каждой строки равны индексу строки
N = int(input())
a = 1
c = -1
while a<=N:
    b = 1
    c = c + 1
    while b<=N:
        c = c
        print(c, end=" ")
        b+=1
    print(" ")
    a+=1
```

```python
# Программа получает на вход натуральное число N (1 ≤ N ≤100). 
# Выведите на экран таблицу размером NxN, заполненную по следующему правилу: элементы четных строк - нули, а нечетных - единицы
n = int(input())
a=1
while a<=n:
    b=1
    while b<=n:
        if a%2!=0:
            c=0
        else:
            c=1
        print(c, end=" ")
        b+=1
    print(" ")
    a+=1
```

```python
# Программа получает на вход натуральное число N (1 ≤ N ≤100). 
# Выведите на экран таблицу размером NxN, заполненную по следующему правилу: 
# строка с индексом 0 и N-1 заполнена 1, все остальные строки 0
n = int(input())
a=0
while a<n:
    b=0
    while b<n:
        if a == 0 or a == n-1:
            c=1
        else:
            c=0
        print(c, end=" ")
        b+=1
    print(" ")
    a+=1
```

```python
# Программа получает на вход натуральное число N (1 ≤ N ≤100). 
# Выведите на экран таблицу размером NxN, заполненную по следующему правилу: ко краям таблицы стоят 1, а в середине 0.
n = int(input())
a=0
while a<n:
    b=0
    while b<n:
        if b == 0 or a == 0 or b == n-1 or a == n-1:
            c=1
        else:
            c=0
        print(c, end=" ")
        b+=1
    print(" ")
    a+=1
```

```python
# Программа получает на вход натуральное число N (1 ≤ N ≤100). 
# Выведите на экран таблицу размером NxN, по углам которой стоят 1, все остальные элементы 0.
n = int(input())
a=0
while a<n:
    b=0
    while b<n:
        if (b == 0 and a == 0) or (b == n-1 and a == n-1) or (b == 0 and a == n-1) or (b == n-1 and a == 0):
            c=1
        else:
            c=0
        print(c, end=" ")
        b+=1
    print(" ")
    a+=1
```

```python
# Программа получает на вход натуральное число N (2 ≤ N ≤100, N - четное). 
# Выведите на экран таблицу размером NxN, верхняя половина которой заполнена 1, а нижняя 0.
n = int(input())
a=0
while a<n:
    b=0
    while b<n:
        if a in range(round(n/2)):
            c=1
        else:
            c=0
        print(c, end=" ")
        b+=1
    print(" ")
    a+=1
```

```python
# Программа получает на вход натуральное число N (2 ≤ N ≤100, N - четное). 
# Выведите на экран таблицу размером NxN, верхняя четверть заполнена 1, а все остальные элементы 0
n = int(input())
a=0
while a<n:
    b=0
    while b<n:
        if a in range(round(n/2)) and b in range(round(n/2)):
            c=1
        else:
            c=0
        print(c, end=" ")
        b+=1
    print(" ")
    a+=1
```

```python
# Программа получает на вход натуральное число N (3 ≤ N ≤100, N - нечетное). 
# Выведите на экран таблицу размером NxN, средний столбец и средняя строка которой заполнены единицами, все остальные элементы нули.
n = int(input())
a=0
while a<n:
    b=0
    while b<n:
        if a == (n-1)/2 or b == (n-1)/2:
            c=1
        else:
            c=0
        print(c, end=" ")
        b+=1
    print(" ")
    a+=1
```

```python
# Программа получает на вход натуральное число N (1 ≤ N ≤100). 
# Выведите на экран таблицу размером NxN, в которой на главной диагонали стоят 1, а все остальные места заполнен 0.
n = int(input())
a=0
while a<n:
    b=0
    while b<n:
        if a==b:
            c=1
        else:
            c=0
        print(c, end=" ")
        b+=1
    print(" ")
    a+=1
```
